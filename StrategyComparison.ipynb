```python
from PriceLoader import PriceLoader, MarketDataPoint

from BenchmarkStrategy import BenchmarkStrategy
from MovingAverageStrategy import MovingAverageStrategy
from VolatilityBreakoutStrategy import VolatilityBreakoutStrategy
from MACDStrategy import MACDStrategy
from RSIStrategy import RSIStrategy

from Engine import StrategyExecutor, Portfolio, plot_portfolio

import pandas as pd
import requests
```

```python
url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
headers = {'User-Agent': 'Mozilla/5.0'}
response = requests.get(url, headers=headers)
tables = pd.read_html(response.text)
sp500_df = tables[0]
sp500_tickers = sp500_df['Symbol'].tolist()
```

```python
benchmark_order_lists = {}
moving_avg_order_lists = {}
volatility_breakout_order_lists = {}
macd_order_lists = {}
rsi_order_lists = {}

for ticker in sp500_tickers:
    pl = PriceLoader(ticker, "2005-01-01", "2025-01-01")
    market_data = pl.fetch_prices()

    if market_data is None or len(market_data) == 0:
        print(f"No market data for {ticker}, skipping.")
        continue
    else:
        benchmark_executor = StrategyExecutor()
        moving_avg_executor = StrategyExecutor()
        volatility_breakout_executor = StrategyExecutor()
        macd_executor = StrategyExecutor()
        rsi_executor = StrategyExecutor()

        benchmark_executor.execute(strategy=BenchmarkStrategy(), data=market_data)
        benchmark_order_lists[ticker] = benchmark_executor.return_orders()

        moving_avg_executor.execute(strategy=MovingAverageStrategy(20,50), data=market_data)
        moving_avg_order_lists[ticker] = moving_avg_executor.return_orders()

        volatility_breakout_executor.execute(strategy=VolatilityBreakoutStrategy(lookback=20), data=market_data)
        volatility_breakout_order_lists[ticker] = volatility_breakout_executor.return_orders()

        macd_executor.execute(strategy=MACDStrategy(), data=market_data)
        macd_order_lists[ticker] = macd_executor.return_orders()

        rsi_executor.execute(strategy=RSIStrategy(), data=market_data)
        rsi_order_lists[ticker] = rsi_executor.return_orders()
```

```python
price_lists = {}

for ticker in sp500_tickers:
    pl = PriceLoader(ticker, "2005-01-01", "2025-01-01")
    market_data = pl.fetch_prices()

    if market_data is None or len(market_data) == 0:
        print(f"No market data for {ticker}, skipping.")
        continue
    else:
        price_lists[ticker] = [data.price for data in market_data]
```

```python
# change the execute_orders method to take in a dictionary of orders and index of the lists
benchmark_portfolio = Portfolio(cash=1000000.0)
benchmark_portfolio.execute_orders(order_lists=benchmark_order_lists, price_lists=price_lists)

moving_avg_portfolio = Portfolio(cash=1000000.0)
moving_avg_portfolio.execute_orders(order_lists=moving_avg_order_lists, price_lists=price_lists)

volatility_breakout_portfolio = Portfolio(cash=1000000.0)
volatility_breakout_portfolio.execute_orders(order_lists=volatility_breakout_order_lists, price_lists=price_lists)

macd_portfolio = Portfolio(cash=1000000.0)
macd_portfolio.execute_orders(order_lists=macd_order_lists, price_lists=price_lists)

rsi_portfolio = Portfolio(cash=1000000.0)
rsi_portfolio.execute_orders(order_lists=rsi_order_lists, price_lists=price_lists)
```
